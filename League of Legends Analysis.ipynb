```python
import numpy as np 
import pandas as pd 
import sklearn
import matplotlib
```

```python
champs = pd.read_csv("champs.csv")
champs.head()
```

```python
matches = pd.read_csv("matches.csv")
matches.head()
```

```python
participants = pd.read_csv('participants.csv')
participants.tail()
```

```python
stats1 = pd.read_csv('stats1.csv')
stats1.head(2)
```

```python
stats2 = pd.read_csv('stats2.csv', low_memory=False)
stats2.head(2)
```

```python
stats = stats1.append(stats2)
stats.shape
```

```python
stats = pd.concat([stats1, stats2])
stats.shape
```

```python
df = (
    participants
    .merge(stats, how='left', on='id')
    .merge(champs, how='left', left_on='championid', right_on='id')
    .merge(matches, how='left', left_on='matchid', right_on='id')
)
```

```python
df.columns
```

```python
df.head(3)
```

```python
def final_position(col):
    if col['role'] in ('DUO_SUPPORT', 'DUO_CARRY'):
        return col['role']
    else:
        return col['position']
```

```python
df['final_position'] = df.apply(final_position, axis=1)
```

```python
df.head(2)
```

```python
df['team_role'] = np.where(df['player'] <= 5, '1', '2') + ' - ' + df['final_position']
```

```python
df['team'] = df['player'].apply(lambda x: '1' if x <= 5 else '2')
df['team_role'] = df['team'] + ' - ' + df['final_position']
```

```python
df.head(10)
```

```python
remove_index = (
    df.groupby(['matchid', 'team_role'])
      .size()  
      .reset_index(name='count')
      .query("count != 1")['matchid']
      .unique()
      .tolist()
)
```

```python
remove_index.extend(df[df['final_position'] == 'BOT']['matchid'].unique())
remove_index = list(set(remove_index))
```

```python
df.columns 
```

```python
df = df[['id', 'matchid', 'player', 'name', 'final_position', 'team_role',
         'win', 'kills', 'deaths', 'assists', 'turretkills','totdmgtochamp',
         'totheal', 'totminionskilled', 'goldspent', 'totdmgtaken', 'inhibkills',
         'pinksbought', 'wardsplaced', 'duration', 'platformid',
         'seasonid']]
df.head()
```
